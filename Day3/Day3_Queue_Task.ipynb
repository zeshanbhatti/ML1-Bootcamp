```python
queue = []

def Enqueue(num):
    if len(queue) <= 10:
        queue.append(num)
        print(f"enqueued {num}")
    else:
        print("queue is full")

def Dequeue():
    if len(queue) > 0:
        removed_item = queue.pop(0)
        print("dequeued item:", removed_item)
    else:
        print("queue is empty, so no-one is removed")

def ViewQueue():
    print("current queue items:", queue)

while True:1
    print("\nMenu:")
    print("1. enqueue")
    print("2. dequeue")
    print("3. view Queue")
    print("4. quit")

    choice = input("enter your choice (1/2/3/4): ")

    if choice == '1':
        num = int(input("enter the number to enqueue: "))
        Enqueue(num)
    elif choice == '2':
        Dequeue()
    elif choice == '3':
        ViewQueue()
    elif choice == '4':
        print("exiting the program")
        break
    else:
        print("invalid choice, please enter 1, 2, 3, or 4")
```
